```python
import asyncio
import aiomysql
import nest_asyncio

# Allow nested event loops in notebooks
nest_asyncio.apply()

async def test_mysql_connection():
    try:
        conn = await aiomysql.connect(
            host='localhost',
            port=3306,
            user='root',
            password='eduflow',
            db='eduflow_db_dev'
        )
        async with conn.cursor() as cur:
            await cur.execute("SELECT NOW();")
            result = await cur.fetchone()
            print("✅ Connected! MySQL server time:", result[0])
        conn.close()
    except Exception as e:
        print("❌ Connection failed:", e)

# Run in notebook
await test_mysql_connection()
```
