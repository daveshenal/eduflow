```python
from datetime import datetime, timezone
import pytz

# Example job ID
job_id = "job-1774547498894"

# Extract timestamp in milliseconds
timestamp_ms = int(job_id.split("-")[1])

# Convert milliseconds to seconds
timestamp_s = timestamp_ms / 1000

# Create UTC datetime
utc_time = datetime.fromtimestamp(timestamp_s, tz=timezone.utc)

# Convert to Sri Lanka timezone
sri_lanka_tz = pytz.timezone("Asia/Colombo")
local_time = utc_time.astimezone(sri_lanka_tz)

# Print results
print("UTC time:", utc_time.strftime("%Y-%m-%d %H:%M:%S"))
print("Sri Lanka time:", local_time.strftime("%Y-%m-%d %H:%M:%S"))
```
