```python
import asyncio
import aiomysql
import ssl

# --- Update these with your server info ---
MYSQL_HOST = ""
MYSQL_PORT = 3306
MYSQL_USER = ""
MYSQL_PASSWORD = ""
MYSQL_DB = ""

# SSL context for Azure MySQL
ssl_context = ssl.create_default_context()

async def test_connection():
    try:
        conn = await aiomysql.connect(
            host=MYSQL_HOST,
            port=MYSQL_PORT,
            user=MYSQL_USER,
            password=MYSQL_PASSWORD,
            db=MYSQL_DB,
            autocommit=True,
            ssl=ssl_context
        )
        async with conn.cursor() as cursor:
            await cursor.execute("SELECT NOW();")
            row = await cursor.fetchone()
            print("✅ Connected! Current DB time:", row[0])
        conn.close()
    except Exception as e:
        print("❌ Connection failed:", e)

# Run the async function in notebook
await test_connection()
```
