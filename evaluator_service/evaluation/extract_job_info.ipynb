```python
import json
import os
from openpyxl import load_workbook

# ==== CONFIG ====
BASE_PATH = r"C:\Users\daves\Desktop\EduFlow\eduflow\evaluator_service\evaluation\downloaded_jobs"
EXCEL_PATH = r"C:\Users\daves\Desktop\EduFlow\eduflow\evaluator_service\evaluation\evaluation_results.xlsx"

JOB_IDS = [
    "job-1774534470161",
    "job-1774534999912",
    "job-1774544649734",
    "job-1774545286253",
    "job-1774545754239",
    "job-1774547498894",
    "job-1774548177351",
    "job-1774548651030",
    "job-1774549026730",
    "job-1774574045744",
    "job-1774574325545",
    "job-1774574626655",
    "job-1774590551422",
    "job-1774584736262",
    "job-1774584969330",
    "job-1774596210500",
    "job-1774595522117",
    "job-1774596683553",
]

# ==== LOAD EXCEL ====
wb = load_workbook(EXCEL_PATH)
ws = wb.active

# Find column indexes
header = [cell.value for cell in ws[1]]
job_id_col       = header.index("Job ID") + 1
input_tokens_col = header.index("Total Input Tokens") + 1
output_tokens_col= header.index("Total Output Tokens") + 1
duration_col     = header.index("Duration") + 1
num_docs_col     = header.index("Number of Docs") + 1
inputs_col       = header.index("Input Parameters") + 1

print(f"Columns found — Job ID: {job_id_col}, Input Tokens: {input_tokens_col}, "
      f"Output Tokens: {output_tokens_col}, Duration: {duration_col}, "
      f"Number of Docs: {num_docs_col}, Inputs: {inputs_col}")

# ==== MAIN LOOP ====
for job_id in JOB_IDS:
    usage_path = os.path.join(BASE_PATH, job_id, "logs", "usage.json")
    input_params_path = os.path.join(BASE_PATH, job_id, "logs", "input_params.json")

    # Load usage.json
    if os.path.exists(usage_path):
        with open(usage_path, "r") as f:
            usage = json.load(f)
        input_tokens  = usage.get("total_input_tokens")
        output_tokens = usage.get("total_output_tokens")
    else:
        print(f"[{job_id}] usage.json not found at {usage_path}")
        input_tokens = None
        output_tokens = None

    # Load input_params.json
    if os.path.exists(input_params_path):
        with open(input_params_path, "r") as f:
            input_params = json.load(f)
        duration     = input_params.get("duration")
        num_docs     = input_params.get("num_docs")
        inputs_json  = json.dumps(input_params)  # store full JSON as string
    else:
        print(f"[{job_id}] input_params.json not found at {input_params_path}")
        duration = None
        num_docs = None
        inputs_json = None

    print(f"[{job_id}] input={input_tokens}, output={output_tokens}, "
          f"duration={duration}, num_docs={num_docs}")

    # Find matching row and update
    updated = False
    for row in ws.iter_rows(min_row=2):
        if row[job_id_col - 1].value == job_id:
            ws.cell(row=row[0].row, column=input_tokens_col,  value=input_tokens)
            ws.cell(row=row[0].row, column=output_tokens_col, value=output_tokens)
            ws.cell(row=row[0].row, column=duration_col,     value=duration)
            ws.cell(row=row[0].row, column=num_docs_col,     value=num_docs)
            ws.cell(row=row[0].row, column=inputs_col,       value=inputs_json)
            updated = True
            break

    if not updated:
        print(f"[{job_id}] Job ID not found in Excel")

# ==== SAVE ====
wb.save(EXCEL_PATH)
print("\nDone! Excel updated.")
```
