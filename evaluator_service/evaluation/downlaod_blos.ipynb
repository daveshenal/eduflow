```python
import os
from azure.storage.blob import BlobServiceClient

CONNECTION_STRING = os.getenv("AZURE_STORAGE_CONNECTION_STRING")

# Container name
CONTAINER_NAME = "ai-saves"

# Prefix root
PREFIX = "index-0001"

# Job IDs you want
job_ids = [
    "job-1774534470161",
    "job-1774534999912",
    "job-1774544649734",
    "job-1774545286253",
    "job-1774545754239",
    "job-1774547498894",
    "job-1774548177351",
    "job-1774548651030",
    "job-1774549026730"
]

# Local base directory
BASE_DIR = "downloaded_jobs"

# Create client
blob_service_client = BlobServiceClient.from_connection_string(CONNECTION_STRING)
container_client = blob_service_client.get_container_client(CONTAINER_NAME)

os.makedirs(BASE_DIR, exist_ok=True)


def download_blob(blob_name):
    # Remove "index-0001/" to keep clean local structure
    relative_path = blob_name.replace(f"{PREFIX}/", "")
    local_path = os.path.join(BASE_DIR, relative_path)

    os.makedirs(os.path.dirname(local_path), exist_ok=True)

    blob_client = container_client.get_blob_client(blob_name)

    with open(local_path, "wb") as f:
        data = blob_client.download_blob()
        f.write(data.readall())

    print(f"Downloaded: {local_path}")


# Loop through each job
for job_id in job_ids:
    print(f"\nProcessing {job_id}...")

    prefix = f"{PREFIX}/{job_id}/"

    blobs = container_client.list_blobs(name_starts_with=prefix)

    found = False

    for blob in blobs:
        found = True
        download_blob(blob.name)

    if not found:
        print(f"No blobs found for {job_id}")
```
